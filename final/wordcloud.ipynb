```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from stylecloud import gen_stylecloud
```

# 1. 对专辑原始文本数据绘制词云图（几张重要专辑，不分析情感）

## Fearless

```python
fearless = pd.read_csv("D:/intro/DaSE_intro/final/lyrics/02-fearless_taylors_version.csv")

# 小写
fearless['clean_lyric'] = pd.Series(fearless['lyric'], dtype = "string")
fearless['clean_lyric'] = fearless['clean_lyric'].str.lower()

stopword = open("D:/intro/DaSE_intro/final/stopwords.txt", 'r').read()
fearless['clean_lyric'] = fearless['clean_lyric'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopword)]))

# 去掉标点
fearless['clean_lyric'] = fearless['clean_lyric'].str.replace('[^\w\s]', '')
fearless['clean_lyric'] = fearless['clean_lyric'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stopword)]))
fearless.head()
```

```python
sentence = [line.strip() for line in fearless['clean_lyric']]
cut_word = " ".join(sentence)

gen_stylecloud(text = cut_word,
              gradient = 'horizontal',
              size = 1024,
              palette = 'cartocolors.qualitative.Antique_10',
              background_color = 'white',
              max_words = 100,
              collocations = False,
              icon_name = 'fas fa-heart',
              output_name = r'D:/intro/DaSE_intro/final/pics/fearless.png')
```
